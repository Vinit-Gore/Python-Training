````python
'''
Given a list of lists, find the list with highest sum of elements.
'''
list1 = [[10,20,30], [40,50,60], [70,80,90],[-2,300,-200]]
sum = [0]*len(list1)
for (indexValue, lo) in enumerate(list1):
  for li in lo:
    sum[indexValue]+=li
i = sum.index(max(sum))  
print(list1[i])
````

````python
'''
Given a list of tuples, find the smallest second index value from a list of tuples.
'''
list2 = [(10,20),(-2,40),(0,0), (10,-2)]
min = 0
for t in list2:
  if t[1] < min:
    min = t[1]
print(min)
````

````python
'''
Speed of tuple and list looping
'''

import time
COUNT = 10000

list1 = list(range(10, COUNT, 10))
tuple1 = tuple(range(10, COUNT, 10))


start = time.time()
for t in tuple1:
    pass
end = time.time()
tupletime = end - start

start = time.time()
for l in list1:
    pass
end = time.time()
listtime = end - start

if tupletime < listtime: 
  print("Tuple looping is faster")
else:
  print("List looping is faster")
````

````python
listDict1 = [{"name":"a","age":22,"salary":24581},{"name":"b"},{"name":"c","age":11},{"name":"d","age":22,"salary":12121}]

sumSal = 0
avgAge = 0
count =0
for dictitr in listDict1:
  if 'salary' in  dictitr:
    sumSal += dictitr['salary']
  if 'age' in dictitr:
    avgAge += dictitr['age']
    count+=1
avgAge /= count
print("AvgAge:", avgAge)

print("sum Salary:", sumSal)
````

````python
'''
Program to convert dictionary into CSV file and read it
'''

import csv
csv_columns = ['No','Name','Country']
dict_data = [
{'No': 1, 'Name': 'Alex', 'Country': 'India'},
{'No': 2, 'Name': 'Ben', 'Country': 'USA'},
{'No': 3, 'Name': 'Shri Ram', 'Country': 'India'},
{'No': 4, 'Name': 'Smith', 'Country': 'USA'},
{'No': 5, 'Name': 'Yuva Raj', 'Country': 'India'},
]
csv_file = "Names.csv"
try:
    with open(csv_file, 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for data in dict_data:
            writer.writerow(data)
except IOError:
    print("I/O error") 

with open(csv_file, 'r') as f:
  print("Using readlines:")
  print(f.readlines())
  f.seek(0)
  print('\n Using for loop:')

  for line in f:
    print(line)

  print("Using csv reader:")
  csv.DictReader(dict_data)
````
