```python
from bs4 import BeautifulSoup
import pandas as pd
import json
import requests
```

```python
response = requests.get('https://www.form.io/how-it-works')
```

```python
soup = BeautifulSoup(response.content, 'html5lib')
```

```python
import pymongo
client = pymongo.MongoClient("localhost", 27017)
db = client.WebScraping
```

```python
db.create_collection('formioHowItWorks')
```

```python
singleSection = soup.find('div',attrs={"class":"content"})
#Insert All Steps
for (index,para) in enumerate(singleSection.findAll('p')):
    db.formioHowItWorks.insert_one({'Step': str(index), 'Action': para.text})
```

```python
#Find All Steps
for item in db.formioHowItWorks.find():
    print(item)
```

```python
#Delete all steps
db.formioHowItWorks.delete_many({})
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import json
import requests
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://www.hackerearth.com/challenges/')
```

```python
# response = requests.get('https://www.hackerearth.com/challenges/')
```

```python
res = driver.execute_script("return document.documentElement.outerHTML")
driver.quit()
print(res)
```

```python
soup = BeautifulSoup(res, 'html5lib')
print(soup)
```

```python
box = soup.find('div', {'class':'upcoming challenge-list'})
print(box)
```

```python
all_hackathons = box.find_all('div',attrs={"class":"challenge-card-modern"})
for (index,hackathon) in enumerate(all_hackathons):
    h_type = hackathon.find('div', {'class': 'challenge-type'}).text.replace('\n','').strip()
    name = hackathon.find('div', {'class': 'challenge-name'}).text.replace('\n','').strip()
    date = hackathon.find('div', {'class': 'date'}).text.replace('\n','').strip()
    
    print(index, h_type+','+name+','+date)
```
